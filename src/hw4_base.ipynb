* Смотрим сколько чтений не удалось приписать ни одному гену:
* `__no_feature ` – столько чтений соответствует участкам генома, где не аннотировано ни одного экзона
* `__ambiguous ` – столько чтений могут принадлежать разным генам
* Более подробно про такие случаи можно посмотреть в описании HTSeq-count - https://htseq.readthedocs.io/en/release_0.11.1/count.html
* Находим общее число чтений, соответствующих хотя бы одному гену
* Для каждого образца приводим эту статистику в отчете на Github


```python
import pandas as pd

names = [
    'SRR3414635',
    'SRR3414636',
    'SRR3414637',
    'SRR3414629',
    'SRR3414630',
    'SRR3414631',
]

df = []

for name in names:
    with open(name + '.counts') as file:
        no_feature = 0
        ambiguous = 0

        for line in (file.readlines()[::-1]):
            line = line[:-1]
            if line.startswith("__ambiguous"):
                ambiguous = int(line.split()[1])
            if line.startswith("__no_feature"):
                no_feature = int(line.split()[1])
                break

        df.append({
            "id": name,
            "__no_feature": no_feature,
            "__ambiguous": ambiguous
        })

df = pd.DataFrame(df)
df
```

* Объединям все файлы .counts по генам в один общий файл `ALL.counts`.
* В качестве заголовка для каждого образца указывать не его SRR id, а c1, c2, c3 для контрольных образцов и r1, r2, r3 для перепрограммированных образцов.
* Загружаем полученный ALL.counts файл в Github репозиторий

```python
samples = list(zip(names, ('c1', 'c2', 'c3', 'r1', 'r2', 'r3')))

counts = []
for sample in samples:
    counts.append(
        pd.read_csv(
            sample[0] + ".counts",
            sep="\t",
            names=[sample[1]]
        )
    )

# cutting 5 last lines
counts_1 = pd.concat(counts, axis=1)
all_counts = counts_1[:-5]
all_counts.to_csv("ALL.counts")

all_counts
```
